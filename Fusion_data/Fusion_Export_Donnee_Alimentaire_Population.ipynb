```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
#import du datafrme data_export_import

data_export_import = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_export_import.csv',decimal=".", index_col=0)
data_export_import.head()
```

```python
#import du datafrme data_dispo_calorie

data_dispo_calorie = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_dispo_calorie.csv',decimal=".", index_col=0)
data_dispo_calorie.head()
```

```python
#import du datafrme df_alimentation_animeaux

df_alimentation_animeaux = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/df_alimentation_animeaux.csv',decimal=".", index_col=0)
df_alimentation_animeaux.head()
```

```python
# jointure des dataframes data_dispo_calorie et df_alimentation_animeaux

data_A = data_dispo_calorie.merge(df_alimentation_animeaux, on=["Code zone","Zone"], how='outer')
display(data_A)
```

```python
# jointure des dataframes data_dispo_calorie + df_alimentation_animeaux + data_export_import

data_B = data_A.merge(data_export_import, on=["Code zone","Zone"], how='outer')
data_B.reset_index(inplace=True)
display(data_B)
```

```python
#import du datafrme poplation_taux_croissance

poplation_taux_croissance = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/poplation_taux_croissance.csv',decimal=".", index_col=0)
poplation_taux_croissance.head()
```

```python
# jointure des dataframes data_dispo_calorie + df_alimentation_animeaux + data_export_import + poplation_taux_croissance

data_C = data_B.merge(poplation_taux_croissance, on=["Code zone","Zone"], how='outer')
display(data_C)
```

```python
#Renommage de la  colonne taux_croissance
data_C = data_C.rename(columns={'taux_croissance':'Taux de croissance_population [%]'})
```

```python
#préparation de l'export du dataframe en fichier CSV

data_C.to_csv('/content/drive/MyDrive/Colab Notebooks/Fusion_Donnee_Alimentaire_Population.csv', index=False)
```
