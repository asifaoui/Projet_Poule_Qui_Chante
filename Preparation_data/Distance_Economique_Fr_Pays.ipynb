```python
# Lecture du fichier Indicateur de la distance entre les pays
#La base de données (dist_cepii) est dyadique, c'est-à-dire qu'elle inclut des variables définies au niveau des paires de pays. La distance géographique est l'exemple le plus classique de ces variables,
#et le fichier inclut différentes mesures de distance bilatérales (en km) pour la plupart des paires de pays du monde.
df_dist_pays = pd.read_excel('/content/drive/MyDrive/Colab_Projet_Poule/dist_cepii.xls')
df_dist_pays.head()
```

```python
# Filtrer France (code ISO3 = 'FRA')
france_dist = df_dist_pays[df_dist_pays['iso_o'] == 'FRA'][['iso_o', 'iso_d', 'dist', 'distcap', 'distw','contig','distwces']]
france_dist.head()
```

```python
france_dist.shape
```

```python
france_dist = france_dist.rename(columns={'iso_d': 'Country Code', 'dist': 'distance_simple',
                                         'distcap': 'distance_capitales', 'distw': 'Distance_ponderee Pays_France(Km)','distwces':'Distance_Economique Pays_France(Km)'})
print(france_dist.head())
```

```python
print(france_dist.loc[france_dist['Country Code']== 'CCK'])
```

```python
#Filtrer pour les variables Country Code et distance_ponderee
var_distance = ['Country Code','Distance_ponderee Pays_France(Km)','Distance_Economique Pays_France(Km)']
france_dist = france_dist[var_distance]
#france_dist.reset_index(inplace=True)
france_dist.head()
```

```python
#vérification des valeurs manquantes
print(france_dist.isnull().sum())
```

```python
#Vérification des Doublons
france_dist.loc[france_dist[['Country Code','Distance_ponderee Pays_France(Km)','Distance_Economique Pays_France(Km)']].duplicated(keep=False),:]
```

```python
#Verification de type des variables
france_dist.dtypes
```

```python
#Conversion du type de la varible d'object à float
france_dist_num = france_dist
france_dist_num['Distance_ponderee Pays_France(Km)'] = pd.to_numeric(france_dist_num['Distance_ponderee Pays_France(Km)'], errors='coerce')
```

```python

france_dist_num['Distance_Economique Pays_France(Km)'] = pd.to_numeric(france_dist_num['Distance_Economique Pays_France(Km)'], errors='coerce')
```
