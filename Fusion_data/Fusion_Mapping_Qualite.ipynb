```python
#import du fichier CSV Mapping_PIB.csv

Mapping_PIB = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Mapping_PIB.csv',decimal=".")

Mapping_PIB.head()
```

```python
#import du fichier CSV Qualite_Reglementaire.csv

data_qualite_reglementaire = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Qualite_Reglementaire.csv',decimal=".")

data_qualite_reglementaire.head()
```

```python
# jointure des 2 dataframes sur  colonnes Country Code

Mapping_Qualite = Mapping_PIB.merge(data_qualite_reglementaire, on=["Country Code"], how='outer')
display(Mapping_Qualite)
```

```python
#Suppression de la colonne Unnamed: 0
Mapping_Qualite = Mapping_Qualite.drop('Unnamed: 0', axis=1)
```

```python
#vérification des valeurs manquantes
print(Mapping_Qualite.isnull().sum())
```

```python
#Suppression de la colonne Zone_y
Mapping_Qualite = Mapping_Qualite.drop('Zone_y', axis=1)
```

```python
# Renommage des colonnes de Zone_x
Mapping_Qualite = Mapping_Qualite.rename(columns={'Zone_x': 'Zone'})
Mapping_Qualite.head()
```

```python
#Renommage des colonnes--> plus propre

Mapping_Qualite = Mapping_Qualite.rename(columns={'qualite_reglementaire':'Qualité Reglementaire [-2,5..+2,5]'})
Mapping_Qualite.head()
```
