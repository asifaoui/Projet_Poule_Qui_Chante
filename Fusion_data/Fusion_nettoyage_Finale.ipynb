```python
#Renommage des colonnes--> plus propre
Mapping_Finale = Mapping_Finale.rename(columns={'Disponibilité alimentaire (Kcal/personne/jour)':'Disponibilité alimentaire en Viande de Volaille (Kcal/personne/jour)'})
```

```python
#Afficher les lignes où toutes les valeurs sont NaN

colonnes = ["Disponibilité alimentaire en Viande de Volaille (Kcal/personne/jour)", "Quantité Aliments_disponible Poules(Milliers tonnes)","Export_Utile(Milliers tonnes)" ,"Taux de croissance_population [%]",
           "Emploi dans la production_vegetale_animale (1000No)","Distance_Economique Pays_France(Km)","Indice Stabilite Politique",
            "Croissance du PIB % annuel","Qualité Reglementaire [-2,5..+2,5]","[%]surfaces agricoles biologiques/total des terres agricoles"]

Mapping_Finale[Mapping_Finale[colonnes].isna().all(axis=1)]

```

```python
#Suppression des colonnes Numeric et Code zone

Mapping_Finale.drop(columns=['Numeric', 'Code zone'], inplace=True)
```

```python
#Ajout d'une colonne pour calculer le nb de NaN dans chaque ligne
Mapping_Finale = Mapping_Finale.assign(nb_nan=Mapping_Finale[colonnes].isna().sum(axis=1))
Mapping_Finale.head()
```

```python
Mapping_Finale.shape
```

```python
# La fréquence des Nan dans notre dataframe

frequences = Mapping_Finale.loc[:, 'nb_nan'].value_counts()

# Convertir la série en DataFrame pour un affichage plus lisible
frequences_df = frequences.reset_index()
frequences_df.columns = ['nb_nan', 'Frequence']

# Afficher le tableau des fréquences
print(frequences_df)
```
