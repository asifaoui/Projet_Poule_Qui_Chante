```python
import re

#Copie du dataframe data_PIB
data_index_PIB = data_PIB.copy()

#détecter les colonnes année (ex : 2000 [YR2000])
year_cols = [c for c in data_index_PIB.columns if re.search(r'\b\d{4}\b', str(c))]
pd.set_option('future.no_silent_downcasting', True)
#normaliser placeholders en NaN (ex: '..', '', 'NA')
data_index_PIB[year_cols] = data_index_PIB[year_cols].replace(['..', '', 'NA', 'na', 'N/A'], np.nan)
```

```python
#Nouvelle variable indice_PIB pour l'année 2024
data_index_PIB['indice_PIB'] = data_index_PIB['2024 [YR2024]']

# Conversion de la colonne indice_PIB en numerique
data_index_PIB['indice_PIB'] = pd.to_numeric(data_index_PIB['indice_PIB'], errors='coerce')

# Arrondi des valeurs de l'indice_PIB à 2 décimales
data_index_PIB['indice_PIB'] = round(data_index_PIB['indice_PIB'], 2)

# Renommage des colonnes
data_index_PIB = data_index_PIB.rename(columns={'Country Name': 'Zone'})

#nouveau dataframe avec 3 variable : Zone, Country Code et indice_PIB
variables_PIB = ['Zone','Country Code','indice_PIB']
data_index_PIB = data_index_PIB[variables_PIB]
display(data_index_PIB)
```
