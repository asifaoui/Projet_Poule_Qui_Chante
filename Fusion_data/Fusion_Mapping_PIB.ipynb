```python
#import du fichier CSV Mapping_Politique.csv

Mapping_Politique = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Mapping_Politique.csv',decimal=".")

Mapping_Politique.head()
```

```python
#import du fichier CSV data_index_PIB.csv

data_index_PIB = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data_index_PIB.csv',decimal=".")

data_index_PIB.head()
```

```python
# jointure des 2 dataframes sur  colonnes Country Code

Mapping_PIB = Mapping_Politique.merge(data_index_PIB, on=["Country Code"], how='outer')
display(Mapping_PIB)
```

```python
#Suppression de la colonne Unnamed: 0
Mapping_PIB = Mapping_PIB.drop('Unnamed: 0', axis=1)
```

```python
#vérification des valeurs manquantes
print(Mapping_PIB.isnull().sum())
```

```python
#Suppression de la colonne Zone_y
Mapping_PIB = Mapping_PIB.drop('Zone_y', axis=1)
```

```python
# Renommage des colonnes de Zone_x
Mapping_PIB = Mapping_PIB.rename(columns={'Zone_x': 'Zone'})
Mapping_PIB.head()
```

```python
#Renommage des colonnes--> plus propre

Mapping_PIB = Mapping_PIB.rename(columns={'indice_stabilite':'Indice Stabilite Politique','indice_PIB':'Croissance du PIB % annuel'})
Mapping_PIB.head()
```
