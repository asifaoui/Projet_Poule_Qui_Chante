```python
#Lecture du fichier Final_mapping_pays.csv

mapping_pays= pd.read_csv('/content/drive/MyDrive/Colab_Projet_Poule/Final_mapping_pays.csv')
mapping_pays.head()
```

```python
# jointure des 2 dataframes sur 1 colonnes Code zone

data_join_FR = mapping_pays.merge(data_D, on=["Code zone"], how='outer')
display(data_join_FR)
```

```python
#vérification des valeurs manquantes
print(data_join_FR.isnull().sum())
```

```python
#Détail des valeurs manquantes pour la variable Zone_x
data_join_FR.loc[data_join_FR['Zone_x'].isnull(),:]
```

```python
print(data_join_FR.loc[data_join_FR['Zone_x']== 'Sudan'])
```

```python
# Serbie-et-Monténégro et Soudan(ex) n'existe plus --> Suppression des lignes
data_join_FR = data_join_FR[data_join_FR['Zone_x'].notna()]
```

```python
#Détail des valeurs manquantes pour la variable Zone_y
data_join_FR.loc[data_join_FR['Zone_y'].isnull(),:]
```

```python
#Suppression de la colonne Zone_y
data_join_FR = data_join_FR.drop('Zone_y', axis=1)
```

```python
# Renommage des colonnes de Zone_x
data_join_FR = data_join_FR.rename(columns={'Zone_x': 'Zone'})
data_join_FR.head()
#Kosovo n'a pas un code zone ISO attribué
```

```python
#vérification des valeurs manquantes
print(data_join_FR.isnull().sum())
```

```python
#préparation de l'export du dataframe en fichier CSV

data_join_FR.to_csv('/content/drive/MyDrive/Colab Notebooks/Fusion_FR_Mapping.csv', index=True)
```
