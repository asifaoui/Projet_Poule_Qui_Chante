```python
#Suppression des lignes ayant Country Name null
data_stability = data_stability.drop(index=[214,215,216,217,218])
```

```python
import re
#Copie du dataframe data_stability
data_political_stability = data_stability.copy()

#détecter les colonnes année (ex : 2000 [YR2000])
year_cols = [c for c in data_political_stability.columns if re.search(r'\b\d{4}\b', str(c))]
pd.set_option('future.no_silent_downcasting', True)
#normaliser placeholders en NaN (ex: '..', '', 'NA')
data_political_stability[year_cols] = data_political_stability[year_cols].replace(['..', '', 'NA', 'na', 'N/A'], np.nan)
```

```python
 #remplissage des valeurs .. en 2023 par la valeur la plus récente
data_political_stability['2023 [YR2023]'] = data_political_stability[year_cols].apply(Parcourir.last_valid_value_from_right, axis=1)
data_political_stability
```

```python
# indice_stabilite qui prend les valeurs de la colonne 2023 [YR2023]
data_political_stability['indice_stabilite'] = data_political_stability['2023 [YR2023]']

#nouveau dataframe avec 3 variable : Country Name, Country Code et indice_stabilite
variables_stability = ['Country Name','Country Code','indice_stabilite']
data_political = data_political_stability[variables_stability]

# Renommage des colonnes
data_political = data_political.rename(columns={'Country Name': 'Zone'})

# Conversion de la colonne indice_stabilie en numerique
data_political['indice_stabilite'] = pd.to_numeric(data_political['indice_stabilite'], errors='coerce')

# Arrondi des valeurs de l'indice de stabilite à 2 décimales
data_political['indice_stabilite'] = round(data_political['indice_stabilite'],2)

data_political.head()
```
