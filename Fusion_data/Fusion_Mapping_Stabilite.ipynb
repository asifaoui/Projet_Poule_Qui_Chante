```python
#import du fichier CSV Mapping_Distance.csv

Mapping_Distance = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Mapping_Distance.csv',decimal=".")

Mapping_Distance.head()
```

```python
#import du fichier CSV Stabilite_Politique.csv

Stabilite_Politique = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Stabilite_Politique.csv',decimal=".")

Stabilite_Politique.head()
```

```python
# jointure des 2 dataframes sur 2 colonnes Country Code

Mapping_Politique = Mapping_Distance.merge(Stabilite_Politique, on=["Country Code"], how='outer')
display(Mapping_Politique)
```

```python
#Suppression de la colonne Unnamed: 0
Mapping_Politique = Mapping_Politique.drop('Unnamed: 0', axis=1)
Mapping_Politique.head()
```

```python
#vérification des valeurs manquantes
print(Mapping_Politique.isnull().sum())
```

```python
#Suppression de la colonne Zone_y
Mapping_Politique = Mapping_Politique.drop('Zone_y', axis=1)
```

```python
# Renommage des colonnes de Zone_x
Mapping_Politique = Mapping_Politique.rename(columns={'Zone_x': 'Zone'})
Mapping_Politique.head()
```
